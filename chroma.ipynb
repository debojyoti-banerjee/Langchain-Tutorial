```python
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEndpointEmbeddings
from langchain_community.document_loaders import PyPDFLoader
```

```python
#Building embedding model
embedding=HuggingFaceEndpointEmbeddings(model="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_store=Chroma(embedding_function=embedding,persist_directory="Chroma_db",collection_name="sample_1")
```

```python
loader=PyPDFLoader("Btech_Project.pdf")
data=loader.load()
```

```python
vector_store.add_documents(data)
```

```python
result=vector_store.similarity_search(query="what is the value of hurst exponent for signal 1",k=2)
print(result)
```

```python
for i in result:
    print(i.page_content)
```
